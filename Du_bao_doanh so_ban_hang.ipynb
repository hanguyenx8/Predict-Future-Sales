##<b>1.2 Dự báo doanh số là gì</b>
- Dự báo doanh số là công việc dự đoán số lượng sản phẩm có khả năng bán được trong một khoảng thời gian tương lai. <br>
- Trong kinh doanh, vấn đề dự báo doanh số bán hàng (DBDSBH) là rất quan trọng, nó giúp cho doanh nghiệp sản xuất đúng số lượng sản phẩm cần thiết vào đúng thời điểm, tạo được lợi thế cho doanh nghiệp trước các đối thủ cạnh tranh thông qua các lợi ích như:
  - DBDSBH giúp xác định khối lượng sản xuất phù hợp với nguồn lực sản xuất của doanh nghiệp như thiết bị, vốn, nhân lực, kho bãi.
  - DBDSBH giúp lãnh đạo doanh nghiệp đưa ra quyết định về việc nên mở rộng nhà máy và thay đổi dây truyền hợp sản xuất hoặc nên chuyển hướng nguồn lực của mình để sản xuất các sản phẩm khác.
  - DBDSBH giúp doanh nghiệp chuẩn bị lịch trình sản xuất, lịch trình mua sắm hoặc đặt hàng sớm vật tư.
  - DBDSBH là căn cứ ra quyết định mức độ quảng cáo, định hướng mục tiêu marketing, mục tiêu sản xuất và các hoạt động kinh doanh khác để cuối cùng là đạt được các mục tiêu về doanh số.
  - DBDSBH giúp doanh nghiệp tập trung được nguồn lực vào các sản phẩm chủ lực, giảm chi phí không đáng có như các chi phí lưu kho, bảo quản khi hàng bị tồn đọng không bán được, chi phí lãi vay, hay chi phí luân chuyển hàng hóa từ nơi thừa sang nơi thiếu...<br>
- Việc dự đoán doanh số là rất quan trọng nhưng cũng rất khó khăn vì có rất nhiều đặc trưng gây ảnh hưởng đến quá trình tính toán, ví dụ như:
  - Đối thủ cạnh tranh: các sản phẩm của đối thủ ảnh hưởng trực tiếp đến doanh số của công ty
  - Thay đổi công nghệ: Với sự tiến bộ của công nghệ, các sản phẩm mới xuất hiện trên thị trường thay đổi sở thích của của người tiêu dùng, đồng thời cũng thay đổi chi phi sản xuất sản phẩm.
  - Các yếu tố về chính trị như chính sách thuế, chính sách hỗ trợ của chính phủ cũng gây ảnh hưởng tới doanh số bán hàng
  - Cuối cùng là các yếu tố ảnh hưởng trực tiếp đến quá trình sản xuất, phân phối sản phẩm, các yếu tố chủ quan của doanh nghiệp: ví dụ như vị trí đặt cửa hàng khác nhau sẽ dẫn tới số lượng hàng hóa bán được là khác nhau. Hoặc cùng một loại hàng nhưng quy cách đóng gói khác nhau, mẫu mã màu sắc khác nhau cũng sẽ dẫn đến doanh số khác nhau. Hoặc việc bùng nổ doanh số trong những ngày diễn ra sự kiện như even, ngày lễ, ngày kỷ niệm... cũng là những đặc trưng có thể đưa vào mô hình dự đoán.

##<b>3.2 Trực quan hóa dữ liệu</b>
Để hiểu rõ hơn về tập dữ liệu, ta trực quan hóa một số thông tin bằng các dạng biểu đồ

```python
# Định dạng ngày tháng
sales['date'] = sales['date'].apply(lambda x: x[6:]+'/'+x[3:5]+'/'+x[:2])
sales['date'] = pd.to_datetime(sales['date'])
# Bổ sung cột Giá trị (Values) bằng cách nhân đơn giá với số hàng bán được trong ngày
sales['Values'] = sales['item_price'] * sales['item_cnt_day']
sales_df = sales[['date', 'item_cnt_day', 'Values']].groupby('date').sum()
# Thêm các thuộc tính day, weakend, month, year
sales_df['day'] = sales_df.index.day
sales_df['weekend'] = sales_df.index.dayofweek
sales_df['month'] = sales_df.index.month
sales_df['quarter'] = sales_df.index.quarter
sales_df['year'] = sales_df.index.year
```

<b>Vẽ biểu đồ tổng doanh số, doanh thu theo ngày</b>

```python
# Tạo hàm vẽ biểu đồ doanh số
def plot_timeseries(ts, title):
    ts.plot(kind='line', figsize = (16,6), color = 'violet')
    mean_sales = ts.mean()
    plt.axhline(mean_sales, color='g', linestyle='dashed', linewidth=3, label="mean = "+str(int(mean_sales)))
    plt.title(title, fontsize=22, color = 'blue')
    plt.xlabel('Date time', fontsize=16)
    plt.ylabel('Sales', fontsize=16)
    plt.legend()
    plt.grid()
    plt.show();
# Trực quan hóa doanh số
plot_timeseries(sales_df['item_cnt_day'], 'Sale by date')
```

```python
# Trực quan hóa doanh thu
plot_timeseries(sales_df['Values'], 'Values by date')
```

<b>Vẽ biểu đồ BOX với doanh số theo tuần, theo tháng và theo năm</b>

```python
def plot_box_timeseries(dataframe, y_col, title):
    time = ['weekend', 'month', 'year']
    fig = plt.figure(figsize=(18, 5))
    layout = (1, 3)
    for i in range(len(time)):
        ax = plt.subplot2grid(layout, (0, i))
        sns.boxplot(data=sales_df, x=time[i], y=y_col, ax = ax)
        ax.set_ylabel(title)
        ax.set_title(f'{title} by {time[i]}')
    plt.show();
plot_box_timeseries(sales_df, 'item_cnt_day', 'Sales')
```

<b>Nhìn vào biểu đồ BOX doanh số ta rút ra một số nhận xét:</b>
* Với biểu đồ **Weekend** ta thấy doanh số cao nhất vào ngày thứ 7 và thấp nhất vào ngày thứ 2 của mỗi tuần, vậy khách hàng có thói quen mua sắm nhiều hơn vào các ngày thứ 7 và chủ nhật
* Với biểu đồ **month**, doanh số tăng cao nhất vào tháng 12, thấp nhất vào các tháng giữa năm, như vậy khách hàng thường mua hàng nhiều hơn vào dịp noel, các kỳ nghỉ lễ tết mùa đông
* Với biểu đồ **year**, có vẻ doanh số của công ty đang có dấu hiệu sụt giảm, tuy năm 2015 mới có dữ liệu đến tháng 10 và 2 tháng cuối năm có doanh số lớn nhất nhưng tổng cả năm khó có thể bằng năm 2012 được, điều này cũng thể hiện rõ từ biểu đồ **Sale by date** phía trên.

```python
# Tạo bảng doanh thu theo tháng đối với từng cửa hàng
df = sales[['date_block_num','shop_id','Values']].groupby(by = ['date_block_num','shop_id']).sum()
df = df.unstack()
df = df.where(df>0,0)
doanh_thu_thang_theo_shop = df.fillna(0)
doanh_thu_thang_theo_shop['Tổng'] = doanh_thu_thang_theo_shop.sum(axis = 1)
# Vẽ biểu đồ doanh thu theo tháng của công ty
doanh_thu_thang_theo_shop['Tổng'].plot(kind='bar', figsize=(15, 7), alpha = 0.5, rot=0, label="Doanh thu tháng")
plt.axhline(doanh_thu_thang_theo_shop['Tổng'].mean(), color='g', linestyle='dashed', linewidth=3, label="Doanh thu trung bình")
plt.title('Doanh thu hằng tháng', fontsize=22, color = 'red')
plt.ticklabel_format(axis = 'y', style = 'plain')
plt.ylabel('Doanh thu (Ruble)', fontsize=16)
plt.xlabel('Số tháng bắt đầu từ T1/2013', fontsize=16)
plt.ylim(2,250000000)
plt.legend()
plt.grid()
plt.show()
```

## **1. Tạo mô hình cơ sở Base model**
Mô hình cơ sở là kết quả ban đầu đưa ra để làm thước đo tham chiếu, đánh giá chất lượng giữa các mô hình một cách trực quan.<br>
Có nhiều mô hình cơ sở khác nhau, như Naive Forecast, EMA, SMA... Đặc điểm chung của các mô hình cơ sở là sự đơn giản, các giá trị dự đoán được tính theo công thức xuất phát từ dữ liệu ban đầu mà không cần trải qua quá trình đào tạo phức tạp.<br>

### **1.1 Mô hình Naive forecast**
Một trong những mô hình cơ sở phổ biến nhất cho dự báo chuỗi thời gian là mô hình Naive (còn được gọi là [naïve forecast](https://otexts.com/fpp3/simple-methods.html#na%C3%AFve-method)).<br>
không cần bất kỳ huấn luyện nào, Tất cả những gì mà mô hình naive làm là sử dụng giá trị bước thời gian trước đó để dự đoán giá trị bước thời gian tiếp theo với công thức:

$$\hat{y}_{t} = y_{t-1}$$ 


```python
# Tạo hàm vẽ biểu đồ dữ liệu chuỗi thời gian
def plot_time_series(ts, label=None):
    bin = np.arange(0,34,1)
    ts.plot(kind = 'line', label=label, xticks = bin, figsize = (16,5))
    plt.xlabel("Tháng")
    plt.ylabel("Doanh số (sp)")
    if label:
        plt.legend(fontsize=14) # làm cho nhãn lớn hơn
    plt.grid(True)
```

```python
# Tạo chuỗi naïve forecast
naive_forecast = ts_month.shift(1)
# Trực quan hóa
plt.figure(figsize=(16, 5))
plot_time_series(ts_month, label="Sales")
plot_time_series(naive_forecast, label="Naive forecast")
```

Sử dụng các thước đo là RMSE để đánh giá mô hình

```python
# Tạo bảng lưu kết quả mô hình Naive_forecast
Naive_forecast_result_df = data.iloc[:,[0,1,2,36,37]].copy()
Naive_forecast_result_df['y33_pred'] = Naive_forecast_result_df[32] 
# Tính điểm mse
Naive_forecast_result_df['mse'] = (Naive_forecast_result_df[32] - Naive_forecast_result_df[33])**2
# Lấy mức trung bình làm điểm số cho mô hình
Naive_forecast_mse = round(Naive_forecast_result_df['mse'].mean(),6)
Naive_forecast_rmse = round(np.sqrt(Naive_forecast_mse),6)
print(f'Điểm số Naive_forecast: rmse: {Naive_forecast_rmse}')
```

Như vậy, kết quả model_0 sẽ là một mức khởi đầu mà ta cần phải vượt qua

### **1.2 Tạo mô hình trung bình trượt đơn SMA (Simple Moving Average)**
Hiểu đơn giản, đường trung bình trượt đơn được tạo bởi các điểm là trung bình cộng của chuỗi trong khoảng thời gian W.<br>

```python
# Tạo chuỗi SMA
SMA8 = ts_month.shift(1).rolling(window=8).mean()
# Trực quan hóa
plt.figure(figsize=(16, 5))
plot_time_series(ts_month, label="Sales")
plot_time_series(SMA8, label="SMA(8)")
```

```python
def sma(w):
    # Tạo bảng lưu kết quả mô hình SMA
    SMA_result = data.iloc[:,4:].copy()
    SMA_result = SMA_result.T.shift(1).rolling(window=w).mean().T
    SMA_result_df = data.iloc[:,[0,1,2,36,37]].copy()
    SMA_result_df['y33_pred'] = SMA_result[33]
    # Tính điểm mse
    SMA_result_df['mse'] = (SMA_result_df['y33_pred'] - SMA_result_df[33])**2
    # Lấy mức trung bình làm điểm số cho mô hình
    SMA_result_mse = round(SMA_result_df['mse'].mean(),6)
    SMA_result_rmse = round(np.sqrt(SMA_result_mse),6)
    return SMA_result_rmse
# Chạy mô hình với khoảng W khác nhau từ 6 đến 12    
sma_kq = list()
sma_rt = dict()
for w in range(6,13):
    sma_rt[w] = sma(w)
    sma_kq.append(sma_rt[w])
    print(f'W = {w}, rmse = {sma_rt[w]}')
print(f'Trung vị = {np.median(sma_kq)}')
```

Ta thấy tại trung vị W=9 có rmse = 1.744606

```python
# Lưu bảng kết quả mô hình
score_result.loc['SMA(9)'] = (sma_rt[9])
```

### **1.3 Mô hình trung bình trượt có trọng số EMA (Exponential Moving Average)**
Đường trung bình trượt có trọng số giống đường trung bình trượt đơn nhưng các thời điểm gần hơn được đánh trọng số lớn hơn, do vậy mô hình có thể nắm bắt tốt hơn chuyển động của xu thế so với SMA.<br>

```python
# Tạo chuỗi EMA
EMA = ts_month.shift(1).ewm(span=8, adjust=False).mean()
# Trực quan hóa
plt.figure(figsize=(16, 5))
plot_time_series(ts_month, label="Sales")
plot_time_series(EMA, label="EMA(8)")
```

```python
def ema(w):
    # Tạo bảng lưu kết quả mô hình EMA
    EMA_result = data.iloc[:,4:].copy()
    EMA_result = EMA_result.T.shift(1).ewm(span=w, adjust=False).mean().T
    EMA_result_df = data.iloc[:,[0,1,2,36,37]].copy()
    EMA_result_df['y33_pred'] = EMA_result[33]
    # Tính điểm mse
    EMA_result_df['mse'] = (EMA_result_df['y33_pred'] - EMA_result_df[33])**2
    # Lấy mức trung bình làm điểm số cho mô hình
    EMA_result_mse = round(EMA_result_df['mse'].mean(),6)
    EMA_result_rmse = round(np.sqrt(EMA_result_mse),6)
    return EMA_result_rmse
# Chạy mô hình với khoảng W khác nhau từ 6 đến 12 
ema_kq = list()
ema_rt = dict()
for w in range(6,13):
    ema_rt[w] = ema(w)
    ema_kq.append(ema_rt[w])
    print(f'W = {w}, rmse = {ema_rt[w]}')
print(f'Trung vị = {np.median(ema_kq)}')
```

Ta thấy tại trung vị W=9 có rmse = 1.677714

```python
# Lưu bảng kết quả mô hình
score_result.loc['EMA(9)'] = (ema_rt[9])
```

## **2. Mô hình học sâu DNN**


Tạo tập dữ liệu dataset từ bộ dữ liệu sale

```python
from IPython.display import display
pd.set_option('max_colwidth', 10)
pd.set_option('max_rows', 20)
pd.set_option('max_columns', 40)

# Tạo dữ liệu dataset
dataset = data.loc[:,0:33].copy()
# Thêm cột index để tiện ghép y_pred với y_true sau mỗi lần chia K_Fold
dataset.insert(0, 'idx', dataset.index.values)    
dataset
```

Do dữ liệu ít, ta sẽ dùng phương pháp <font color = red><b>K-Fold (k=10) </b></font> để chia bộ train/test theo tỷ lệ 9/1

```python
# Tạo bộ chia KFold
kfold = KFold(n_splits=10, shuffle=True, random_state = 47)
# Chia dữ liệu input và dữ liệu nhãn
dataset = dataset.values
inputs = dataset[:,:-1]
labels = dataset[:,-1]
```

Tạo hàm train model DNN

```python
# Tạo hàm fit model DNN
def DNN_fit(X_train, y_train, X_val, y_val):
    DNN_model = Sequential()
    DNN_model.add(Dense(128, activation="relu"))
    DNN_model.add(Dense(64, activation="relu"))
    DNN_model.add(Dense(16, activation="relu"))
    DNN_model.add(Dense(1, activation="linear"))
    DNN_model.compile(optimizer='adam', loss='mse',
                  metrics=[tf.keras.metrics.RootMeanSquaredError(name='rmse')])
    # fit model
    DNN_model.fit(X_train, y_train, epochs=250, verbose=0,
              validation_data=(X_val, y_val),
              callbacks=[tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=30, restore_best_weights=True)])
    #mse, rmse = DNN_model.evaluate(X_test, y_test)
    return DNN_model
```

```python
# Chạy mô hình DNN qua các Fold
fold_no = 1
DNN_rmse = dict()
DNN_predict_df = pd.DataFrame(data = dataset[:,1:], index = dataset[:,0])
DNN_predict_df['y_pred'] = None
for train, test in kfold.split(inputs, labels):
    print('----------------------------------------------------------------------------------')
    print(f'Training for fold {fold_no} ...')
    X_train_idx, X_val_idx, y_train, y_val = train_test_split(inputs[train], labels[train], test_size =0.2, random_state=47)
    X_test_idx, y_test = inputs[test], labels[test]
    X_train, X_val, X_test = X_train_idx[:, 1:], X_val_idx[:, 1:], X_test_idx[:, 1:]
    DNN_model = DNN_fit(X_train, y_train, X_val, y_val)
    mse, rmse = DNN_model.evaluate(X_test, y_test)
    DNN_rmse[fold_no] = rmse
    print(f'RMSE_per_fold {fold_no}= {rmse}')
    # Lưu kết quả dự đoán vào bảng DNN_predict_df    
    idx_y_pred = X_test_idx[:,0]
    y_pred = pd.Series(DNN_model.predict(X_test).reshape(-1,), idx_y_pred)
    DNN_predict_df.loc[idx_y_pred,'y_pred'] = y_pred
    fold_no += 1
```

```python
# Trực quan hóa kết quả thành dạng biểu đồ cột
DNN_result = pd.Series(DNN_rmse)
#print(DNN_result.describe())
DNN_result.plot(kind = 'bar', figsize = (12,5), label = 'DNN RMSE', rot = 0)
plt.axhline(DNN_result.mean(), color='orange', linestyle='dashed', linewidth=3, label = f'mean = {round(DNN_result.mean(),5)}')
plt.title(f'Điểm số RMSE của các K-Fold của mô hình DNN', fontsize = 16, color = 'red')
plt.legend(loc = 'lower left');
```

```python
# Lưu bảng kết quả mô hình DNN_multi_series_with_K-Fold
score_result.loc['DNN'] = DNN_result.mean()
```

## **3. Mô hình LSTM**

Tạo hàm train mô hình LSTM

```python
# Tạo hàm fit model LSTM_multi
def LSTM_fit(X_train, y_train, X_val, y_val):
    X_train = np.expand_dims(X_train, axis=1)
    X_val = np.expand_dims(X_val, axis=1)
    LSTM_model_fit = Sequential()
    LSTM_model_fit.add(Bidirectional(                   
                          LSTM(128, activation="relu",
                                return_sequences=True,
                                input_shape=(1,33))))
    LSTM_model_fit.add(LSTM(64, activation="relu"))
    #LSTM_model_fit.add(Dense(32, activation="relu"))
    #LSTM_model_fit.add(Dense(16, activation="relu"))
    LSTM_model_fit.add(Dense(1, activation="linear")) 
    LSTM_model_fit.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=0.00005), loss='mse',
                  metrics=[tf.keras.metrics.RootMeanSquaredError(name='rmse')])
    # fit model
    LSTM_model_fit.fit(X_train, y_train, epochs=50, verbose=0,
              validation_data=(X_val, y_val),
              callbacks=[tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)])
    return LSTM_model_fit
```

Chạy mô hình LSTM lần lượt qua các Fold

```python
fold_no = 1
LSTM_rmse = dict()
LSTM_predict_df = pd.DataFrame(data = dataset[:,1:], index = dataset[:,0])
LSTM_predict_df['y_pred'] = None
# Chạy vòng lặp qua các Fold để lấy kết quả
for train, test in kfold.split(inputs, labels):
    print('------------------------------------------------------------------------')
    print(f'Training for fold {fold_no} ...')
    X_train_idx, X_val_idx, y_train, y_val = train_test_split(inputs[train], labels[train], test_size =0.2, random_state=47)
    X_test_idx, y_test = inputs[test], labels[test]
    X_train, X_val = X_train_idx[:, 1:], X_val_idx[:, 1:]
    X_test =  np.expand_dims(X_test_idx[:, 1:], axis=1)
    LSTM_model = LSTM_fit(X_train, y_train, X_val, y_val)
    mse, rmse = LSTM_model.evaluate(X_test, y_test)
    LSTM_rmse[fold_no] = rmse
    #print(f'LSTM_per_fold {fold_no}= {rmse}')
    # Lưu kết quả dự đoán vào bảng LSTM_predict_df   
    idx_y_pred = X_test_idx[:,0]
    y_pred = pd.Series(LSTM_model.predict(X_test).reshape(-1,), idx_y_pred)
    LSTM_predict_df.loc[idx_y_pred,'y_pred'] = y_pred
    fold_no += 1
```

```python
LSTM_result = pd.Series(LSTM_rmse)
#print(LSTM_result.describe())
LSTM_result.plot(kind = 'bar', figsize = (12,5), label = 'LSTM RMSE', rot = 0)
plt.axhline(LSTM_result.mean(), color='orange', linestyle='dashed', linewidth=3, label = f'mean = {round(LSTM_result.mean(),5)}')
plt.title(f'Điểm số RMSE của các K-Fold của mô hình LSTM', fontsize = 16, color = 'red')
plt.legend(loc = 'lower left');
```

```python
# Tạo dữ liệu data_xgb
data1_xgb = data.copy()
data1_xgb["ID"] = data1_xgb["ID"].astype("str")
data1_xgb["shop_id"] = data1_xgb["shop_id"].astype("str")
data1_xgb["item_id"] = data1_xgb["item_id"].astype("str")
data1_xgb["item_category_id"] = data1_xgb["item_category_id"].astype("str")
data1_xgb = data1_xgb.drop("ID", axis = 1)
data1_xgb = data1_xgb.drop("item_category_id", axis = 1)
data1_xgb = data1_xgb.drop("shop_id", axis = 1)
data1_xgb = data1_xgb.drop("item_id", axis = 1)
data1_xgb
```

```python
# Chia tách tập train và test
features_name = data1_xgb.columns.values
X = data1_xgb[features_name[:-1]]
y = data1_xgb[features_name[-1:]]
```

```python
# Tinh chỉnh tham số cho param
param = {'max_depth': 26,               # Độ sâu tối đa của cây, default=6
        'obj':'gbtree',                 # Loại thuật toán tăng cường
        'subsample':1,                  # Tỷ lệ mẫu phụ của các trường hợp đào tạo (Tỷ lệ lấy dữ liệu đào tạo trước khi trồng cây)
        'min_child_weight': 1,          # Số lượng cá thể tối thiểu cần có trong mỗi nút
        'eta':0.31,                     # tương tự learning_rate: là tốc độ học, default=0.3
        'num_round':1000,               # Số lượng vòng lặp
        'eval_metric':'rmse',           # Thước đo rmse cho hồi quy và logloss để phân loại
        'n_estimators': 10}             # n_estimators : Số lượng trees 
```

```python
# Chạy mô hình XGB qua các K-Fold
XGB_predict_df = pd.DataFrame(data = dataset[:,1:], index = dataset[:,0])
XGB_predict_df['y_pred'] = None
XGB_rmse = dict()
fold_no = 1
for train, test in kfold.split(X, y):
    print('------------------------------------------------------------------------')
    print(f'Training for fold {fold_no} ...')
    id_train, id_val = train_test_split(train, test_size =0.2, random_state=42)
    X_train, y_train = X.iloc[id_train], y.iloc[id_train]
    X_val, y_val = X.iloc[id_val], y.iloc[id_val]
    X_test, y_test = X.iloc[test], y.iloc[test]
    np.random.seed(seed=42)
    # Tạo và train mô hình
    train_matrix = xgb.DMatrix(X_train.values, y_train.values)
    val_matrix  = xgb.DMatrix(X_val.values,  y_val.values)
    evallist = [(train_matrix, 'train'), (val_matrix, 'val')]
    xgb_model = xgb.train(param, train_matrix, evals=evallist, early_stopping_rounds=20)
    # Dự đoán kết quả tập test
    y_true = pd.Series(y_test.values.reshape(-1,))
    predictions = xgb_model.predict(xgb.DMatrix(X_test.values))
    predictions = list(map(lambda x: max(0,x), list(predictions)))
    y_pred = pd.Series(predictions)
    # Tính toán điểm số
    xgb_mae = np.mean(y_true - predictions)
    xgb_mse = np.mean((y_true- y_pred)**2)
    xgb_rmse = round(np.sqrt(xgb_mse),6)
    XGB_rmse[fold_no] = xgb_rmse
    # Lưu kết quả dự đoán vào bảng LSTM_predict_df
    y_pred_series = pd.Series(y_pred.values, X_test.index.values)
    XGB_predict_df.loc[X_test.index.values,'y_pred'] = y_pred_series
    print(f'RMSE_per_fold {fold_no}= {xgb_rmse}')
    fold_no += 1
```

```python
XGB_result = pd.Series(XGB_rmse)
#print(XGB_result.describe())
XGB_result.plot(kind = 'bar', figsize = (12,5), label = 'XGB RMSE', rot = 0)
plt.axhline(XGB_result.mean(), color='orange', linestyle='dashed', linewidth=3, label = f'mean = {round(XGB_result.mean(),5)}')
plt.title(f'Điểm số RMSE của các K-Fold của mô hình XGB', fontsize = 16, color = 'red')
plt.legend(loc = 'lower left');
```

```python
# Tạo hàm để tách chuỗi dưới dạng window
def make_windows(x, window_size=3, horizon=1):
    # 1. Tạo một window có window_size cụ thể
    window_step = np.expand_dims(np.arange(window_size+horizon), axis=0)
    # 2. Tạo mảng 2D gồm nhiều bước window (trừ 1 để tính đến 0 indexing)
    window_indexes = window_step + np.expand_dims(np.arange(len(x)-(window_size+horizon-1)), axis=0).T
    # 3. Lập chỉ mục trên mảng đích
    windowed_array = x[window_indexes]
    # 4. Lấy các window được gắn nhãn
    windows = windowed_array[:, :-horizon]
    labels = windowed_array[:, -horizon:]
    return (windows, labels)
```

```python
# Tạo hàm chạy mô hình với đầu vào WS khác nhau
def DNN_single_fit(time_series, horizon = 1):
    WS_lst = [4,5,6,8,9,10,12,13,15,18,24,30]
    predict_with_WS = dict()
    for WS in WS_lst:
        # Định dạng dữ liệu Windowing
        data = make_windows(time_series, window_size = WS, horizon = horizon)
        X, y = data[0], data[1]
        # Tách chuỗi time series thành tập train, val và test
        split_size = int(len(X)*0.8)
        X_train, y_train = X[:split_size], y[:split_size]
        X_val, y_val = X[split_size:], y[split_size:]
        X_test = time_series[-WS:].reshape(-1, WS)
        # Thiết lập random seed
        tf.random.set_seed(42)
        # Xây dựng mô hình
        model_DNN = tf.keras.Sequential([
            Dense(64, activation="relu"),
            Dense(32, activation="relu"),
            Dense(1, activation="linear")])
        # Biên dịch mô hình
        model_DNN.compile(loss="mae",
                        optimizer=tf.keras.optimizers.Adam(),
                        metrics=[ tf.keras.metrics.RootMeanSquaredError(name='rmse')])
        # Khớp mô hình 
        model_DNN.fit(x=X_train, y=y_train, 
                    epochs=500, verbose=0, batch_size=1,
                    validation_data=(X_val, y_val),
                    callbacks=[tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=100, restore_best_weights=True)])
        # Kiểm tra điểm số
        score_test = model_DNN.evaluate(X_val, y_val, verbose=0)
        y_pred = model_DNN.predict(X_test, verbose=0)
        y_pred = round(float(y_pred),6)
        predict_with_WS[WS] = y_pred
        print(f'WS = {WS}, mse = {score_test[0]}, y_pred = {y_pred}')
    return predict_with_WS
```

```python
id = 1258
row_series = data.loc[id,0:].astype(np.float16).values
DNN_predict_series_with_WS = DNN_single_fit(row_series)
# Vẽ biểu đồ minh họa kết quả
item_id	 = data.loc[id,'item_id']
shop_id = data.loc[id,'shop_id']
plot_time_series(data.loc[id,0:], label='Sale')
for WS, value in DNN_predict_series_with_WS.items():
    plot_time_series(pd.Series(data=[data.loc[id,32], value], index=[32,33]), label=f'WS={WS}')
plt.title(f'Biểu đồ doanh số của mặt hàng id={item_id}, shop_id={shop_id}', fontsize = 18, color = 'red')
plt.legend();
```
